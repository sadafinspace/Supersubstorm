```python
import pyspedas
import pytplot
from pytplot import tplot, options, tlimit, tplot_names
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib.colors import ListedColormap
from mpl_toolkits.axes_grid1 import make_axes_locatable
import os
import datetime
import netCDF4
from netCDF4 import Dataset, num2date
import datetime as dt
```

## Probe (a)

```python
pyspedas.rbsp.mageis(trange=['2015-09-07/14:00', '2015-09-07/18:00'], probe = 'a' , level='l3', rel='rel04', time_clip = True)
```

```python
from pytplot import get_data

data_a = get_data('FEDU')
```

```python
data_a.v2
```

```python
## Line plot (33-143KeV)

enspec_at_pitch90_a1 = data_a.y[:, 5, 0:5]
from pytplot import store_data, options, tplot

store_data('en_at_pitch_90_a1', data={'x': data_a.times, 'y': enspec_at_pitch90_a1, 'v': data_a.v2})

# options('en_at_pitch_90_a', 'spec', True)
# options('en_at_pitch_90_a', 'zlog', True)
options('en_at_pitch_90_a1', 'x_interp', True)
options('en_at_pitch_90_a1', 'y_interp', True)
options('en_at_pitch_90_a1', 'ylog', True)

options('en_at_pitch_90_a1', 'ytitle', 'Electron Flux (probe a)')
options('en_at_pitch_90_a1', 'ztitle', 'FEDU-Electron Flux($cm^{-2}$$s^{-1}$$sr^{-1}$$eV^{-1}$)')

options('en_at_pitch_90_a1', 'legend_names', ['33 KeV', '54 KeV', '80 KeV', '108 KeV', '143 KeV'])
options('en_at_pitch_90_a1', 'thick', 1)
pytplot.timebar(1441642020.0, thick=3, color='red')

# options('en_at_pitch_90_a', 'yrange', [1.8, 7.7])

# tplot('en_at_pitch_90_a')
# Add title
# pytplot.tplot('en_at_pitch_90_a1')

# # Show plot
# pytplot.tplot()

pytplot.tplot("en_at_pitch_90_a1", var_label = ["L", "MLT" ])

```
