```python
filename = "/Users/sians/OneDrive/Desktop/LANL_2015/Data file/20150907_LANL-04A_10sec_flux.txt"

# Open the file in read mode
with open(filename, "r") as file:
    # Read the first 130 lines into a string array
    lanl_header = []
    for i in range(130):
        line = file.readline()
        lanl_header.append(line.strip())

    # Read the remaining data into a floating-point array
    data_array = []
    for line in file:
        values = line.strip().split()  # Split the line into individual values

        # Convert the ISO timestamp in the first column to Unix timestamp
        iso_timestamp = values[0]
        dt = datetime.datetime.fromisoformat(iso_timestamp[:-1])
        unix_timestamp = dt.timestamp()

        # Replace the ISO timestamp with the Unix timestamp in the first column
        values[0] = unix_timestamp

        # Convert the remaining values to float
        float_values = [float(x) for x in values[1:]]
        
        # Add the Unix timestamp and the float values to the float array
        data_array.append([unix_timestamp] + float_values)
```

```python
import datetime as dt
time = [row[0] for row in data_array] # first coulmn
# print(timefrac)
lat = [row[1] for row in data_array] #second column
long = [row[3] for row in data_array] #fourth column

#converting time to datetime
time_new=[]
for i in range(0,len(time)):
    time_new= np.append(time_new, datetime.datetime.utcfromtimestamp(time[i]))
```

```python
import datetime

def calculate_mlt(utc_list, longitude_list):
    mlt_list = []
    for utc, longitude in zip(utc_list, longitude_list):
        ut_hour = utc.hour + (utc.minute / 60)  # Calculate the UT hour (including the minute component)
        mlt = (ut_hour + (longitude / 15)) % 24  # Calculate the MLT
        mlt_list.append(mlt)
    return mlt_list

# Example usage for LANL mission
utc_list = time_new[:] 
longitude_list = long[:]  # Example longitudes for LANL mission

mlt_result = calculate_mlt(utc_list, longitude_list)
print(mlt_result)
```

```python
# Create a NumPy array containing the latitude, longitude, and MLT values
data = np.array([lat, longitude_list, mlt_result]).T

# Create a table with column headers
table = [['Latitude', 'Longitude', 'MLT']]
table.extend(data)
```

```python
# Save MLT, latitude, and longitude values to a CSV file
csv_file = 'mlt_coordinates_04A.csv'

rows = zip(utc_list, longitude_list, lat, mlt_result)

with open(csv_file, 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['UTC', 'Longitude', 'Latitude', 'MLT'])  # Write header row
    writer.writerows(rows)

print("MLT, latitude, and longitude values saved to", csv_file)
```
