## Probe A

```python
pyspedas.rbsp.rept(trange=['2015-09-07/12:00', '2015-09-07/19:00'], probe = 'a', time_clip = True)

from pytplot import get_data

data_a = get_data('FEDU')
```

```python
# Energy spectogram at pitch angle 90
from pytplot import store_data, options, tplot

en_spec_at_pa90_a = data_a.y[:, 8, :]

store_data('enspec_at_pa90_a', data={'x': data_a.times, 'y': en_spec_at_pa90_a, 'v': data_a.v2})

options('enspec_at_pa90_a', 'spec', True)
options('enspec_at_pa90_a', 'zlog', True)
options('enspec_at_pa90_a', 'x_interp', True)
options('enspec_at_pa90_a', 'y_interp', True)

# tplot('pad_at_1.8MeV_a')
# pytplot.timebar(1441642020.0, thick=3, color='red')
# pytplot.timebar(1441639200.0, thick=3, color='magenta')
# pytplot.timebar(1441640700.0, thick=3, color='magenta')


pytplot.tplot("enspec_at_pa90_a", var_label = ["L", "MLT" ])
```

```python
## Line plot.

enspec_at_pitch90_a1 = data_a.y[:, 8, 0:6]
from pytplot import store_data, options, tplot

store_data('en_at_pitch_90_a1', data={'x': data_a.times, 'y': enspec_at_pitch90_a1, 'v': data_a.v2})

options('en_at_pitch_90_a', 'spec', True)
options('en_at_pitch_90_a', 'zlog', True)
options('en_at_pitch_90_a1', 'x_interp', True)
options('en_at_pitch_90_a1', 'y_interp', True)
options('en_at_pitch_90_a1', 'ylog', True)

options('en_at_pitch_90_a1', 'ytitle', 'Electron Flux (probe a)')
options('en_at_pitch_90_a1', 'ztitle', 'FEDU-Electron Flux($cm^{-2}$$s^{-1}$$sr^{-1}$$eV^{-1}$)')

options('en_at_pitch_90_a1', 'legend_names', ['1.8 MeV' ,  '2.1 MeV',  '2.6 MeV',  '3.4 MeV',  '4.2 MeV', '5.2 MeV'])
options('en_at_pitch_90_a1', 'thick', 1)
pytplot.timebar(1441642020.0, thick=3, color='red')

# options('en_at_pitch_90_a', 'yrange', [1.8, 7.7])

# tplot('en_at_pitch_90_a')
# Add title
# pytplot.tplot('en_at_pitch_90_a1')

# # Show plot
# pytplot.tplot() 

pytplot.tplot("en_at_pitch_90_a1", var_label = ["L", "MLT" ])

```
